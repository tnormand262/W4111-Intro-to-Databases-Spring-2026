## Q5

#### Question

What is the key difference between a _unique_ constraint and a _primary key_ constraint?

#### Answer

Both offer to uniquely identify a tuple (row) but a primary key is a minimal candidate key. That means that a unique constraint is a candidate key that can be null, whereas a primary key cannot be null. Lastly, unique allows for multiple null attributes (null <> null) whereas primary key constraints prohibit it entirely. Thus, the primary key constraint implicitly enforces the not null constraint

## Q6

#### Question

Briefly explain the concepts of _disjoint/overlapping_ and _total/partial_ with respect to inheritance.

#### Answer

1. Disjoint: an entity can belong to only one subtype ex. a vehicle can be a subtype of truck or car but not both
2. Overlapping: an entity can belong to more than one subtype ex. a person could be an employee and a student at the same time
3. Total: Every entity in the supertype must appear in at least one subtype.
4. Partial: An entity in the supertype may or may not belong to any subtype.


Together, we can have 4 constraint pairs that each deal with how an entity belongs to subtypes.

Disjoint & partial: an entity can belong to 1 or 0 subtypes

Disjoint & total: an entity can belong to exactly 1 subtype

Overlapping & partial: an entity can belong 0 or many subtypes

Overlapping & total: an entity can belong to at least 1 (or more) subtypes


## Q7

#### Question

Briefly explain the following concepts:
1. Natural join
2. Theta join
3. Inner join
4. Outer join

#### Answer

1. Natural join: joins two or more relations on identically named attributes with same attribute domain and values, can cause issues when attributes have same names but diff meanings
2. Theta join: joins relations on a predicate theta. Avoids the issue of natural join joining on unrelated attributes
3. Inner join: joins two relations and returns only the rows that have matching values in both relations on the join condition. Non-matching rows are discarded entirely. This is the default JOIN behavior in SQL. Sort of like full outer join with no null values
4. Outer join: can be obtained by composing a left outer join with a right outer join (or vice versa) and preserves all rows from one or both relations, filling in NULL for missing values where there is no match. 
Comes in three varieties: left outer join: keeps all rows from the left relation, right outer join (similar to left outer join), and full outer join (keeps all rows from both relations)

Users can use the same query language to access the catalog which they use to access the database itself._

What is the name of the schema/database for the catalog in MySQL?

#### Answer

## ER2

#### Question

You are to translate/write DDL to create an SQL database implementing the following diagram. Unless otherwise specified, you can assume that all columns are ```VARCHAR(64).```

The model has the additional constraints:
1. All ```email``` columns are ```unique``` within the table.
2. ```School.school_code``` is always exactly 4 characters.
3. ```Department.department_code``` is always exactly 4 characters.
4. ```CUID``` is ```unique``` and ```not null```.
5. ```Course.faculty_code``` and ```Section.course_code``` are a single characters and the allowed values are "C", "B", "E" and "W".
6. ```course_no``` is always exactly 4 characters and only the digits "0", "1", ..., "9" are allowed characters.
7. The allowed values for ```Section.semester``` are ```1```, ```2``` and ```3```.
8. ```StudentSection.role``` can have the values ```Enrolled```, ```Waitlist```, ```TA```.

#### Answer

```````python
%%sql

drop database if  exists w4111_hw2;
create database w4111_hw2;
USE w4111_hw2;

CREATE TABLE school(
    school_code char(4) PRIMARY KEY,
    school_name varchar(64)
);

CREATE TABLE student(
    UNI VARCHAR(64) PRIMARY KEY,
    last_name VARCHAR(64),
    first_name VARCHAR(64),
    middle_name VARCHAR(64),
    email VARCHAR(64),
    school_code char(4),
    CUID VARCHAR(64) not NULL,
    UNIQUE(email),
    UNIQUE(CUID),
    FOREIGN KEY (school_code) REFERENCES school(school_code)
);

CREATE TABLE Department(
  school_code char(4),
  department_code char(4) PRIMARY KEY,
  department_name VARCHAR(64),
  FOREIGN KEY (school_code) REFERENCES school(school_code)
);

CREATE TABLE Faculty(
  UNI VARCHAR(64) PRIMARY KEY,
  last_name VARCHAR(64),
  first_name VARCHAR(64),
  middle_name VARCHAR(64),
  email VARCHAR(64),
  department_code char(4),
  FOREIGN KEY (department_code) REFERENCES Department(department_code),
  UNIQUE(email)
);

CREATE TABLE Building(
  building_code VARCHAR(64) PRIMARY KEY,
  building_name VARCHAR(64)
);

CREATE TABLE Room(
  room_id VARCHAR(64) PRIMARY KEY,
  building_code VARCHAR(64),
  room_no VARCHAR(64),
  capacity VARCHAR(64),
  FOREIGN KEY (building_code) REFERENCES Building(building_code)
);

CREATE TABLE Course(
  department_code char(4),
  faculty_code CHAR(1),
  course_no CHAR(4),
  course_title VARCHAR(64),
  course_description VARCHAR(64),
  
  PRIMARY KEY (department_code, faculty_code, course_no),
  FOREIGN KEY (department_code) REFERENCES Department(department_code),
  CHECK (faculty_code IN ('C', 'B', 'E', 'W')),
  CHECK ( 
      SUBSTRING(course_no, 1, 1) BETWEEN '0' AND '9' AND
      SUBSTRING(course_no, 2, 1) BETWEEN '0' AND '9' AND
      SUBSTRING(course_no, 3, 1) BETWEEN '0' AND '9' AND
      SUBSTRING(course_no, 4, 1) BETWEEN '0' AND '9'
  )
);


CREATE TABLE section(
  callno VARCHAR(64) PRIMARY KEY,
  department_code CHAR(4), 
  faculty_code CHAR(1),
  course_no char(4),
  section_no VARCHAR(64),
  semester CHAR(1),
  year VARCHAR(64),
  time_slot_id VARCHAR(64), 
  room_id VARCHAR(64),
  instructor_uni VARCHAR(64),
  
  CHECK (
    SUBSTRING(course_no, 1, 1) BETWEEN '0' AND '9' AND
    SUBSTRING(course_no, 2, 1) BETWEEN '0' AND '9' AND
    SUBSTRING(course_no, 3, 1) BETWEEN '0' AND '9' AND
    SUBSTRING(course_no, 4, 1) BETWEEN '0' AND '9'
),
  CHECK(
    faculty_code in ('C','B','E','W')
  ),
  CHECK(
    semester in ('1','2','3')
  ),
  
  FOREIGN KEY (department_code,faculty_code,course_no) REFERENCES Course(department_code, faculty_code,course_no),
  FOREIGN KEY (instructor_uni) REFERENCES Faculty(UNI),
  FOREIGN KEY (room_id) REFERENCES Room(room_id)
);

CREATE TABLE StudentSection(
  UNI VARCHAR(64),
  callno VARCHAR(64) ,
  role VARCHAR(64) ,
  status VARCHAR(64),
  PRIMARY KEY (UNI, callno,role),
  FOREIGN KEY (UNI) REFERENCES student(UNI),
  FOREIGN KEY (callno) REFERENCES section(callno),
  CHECK (role in ('Enrolled', 'Waitlist', 'TA'))
);
```````

## S1

#### Question



Write and execute the SQL query that is equivalent to the relational algebra expression for question R1.

```π ID,name(instructor) - π ID,name(σ instructor.ID = advisor.i_id (instructor ⨯ advisor))```


#### Answer

```````python
%%sql
use db_book;
SELECT ID,name 
FROM instructor
EXCEPT
    SELECT ID,name 
    FROM instructor
    JOIN advisor
    ON instructor.ID = advisor.i_id
;
```````

## S2

#### Question



Write and execute the SQL query that is equivalent to the relational algebra expression for question R2.

```π instructor.ID, instructor.name, course.course_id, course.title,course.credits(σ instructor.dept_name='Comp. Sci.'(```
```(instructor ⨝ instructor.ID = teaches.ID teaches) ⨝ teaches.course_id = course.course_id course))```

#### Answer

```````python
%%sql
SELECT DISTINCT instructor.ID, instructor.name, course.course_id, course.title,course.credits 
FROM instructor
JOIN teaches
    ON instructor.ID = teaches.ID
JOIN course
    ON teaches.course_id = course.course_id
WHERE 
    instructor.dept_name = 'Comp. Sci.'
;
```````

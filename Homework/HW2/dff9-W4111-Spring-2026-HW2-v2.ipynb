Customer

Attributes

- customerNumber — unique identifier for a customer  
- customerName

- addressLine1  
- addressLine2  
- city  
- state  
- postalCode  
- country  
- creditLimit  

Business Rules

- Each customer has a unique `customerNumber`.  
- A customer may place zero, one, or many orders.  
- A customer may make zero, one, or many payments.  

---

Orders

Each **Order** represents a purchase transaction placed by a customer.

Attributes

- orderNumber — unique identifier for an order  
- orderDate  
- requiredDate  
- shippedDate  
- status  
- comments  

Business Rules

- Each order has a unique `orderNumber`.  
- Each order is placed by exactly one customer.  
- An order must contain at least one product.  
- An order may contain many products.  

---

OrderDetails

The **OrderDetails** table represents the individual line items of an order.

Attributes

- orderNumber  
- productCode  
- quantityOrdered  
- priceEach  
- orderLineNumber  

Business Rules

- Each row represents one product appearing on one order.  
- The combination of `orderNumber` and `productCode` uniquely identifies a line item.  
- Each line item belongs to exactly one order.  
- Each line item references exactly one product.  

---

Products

Each **Product** represents an item available for sale.

Attributes

- productCode — unique identifier for a product  
- productName  
- productLine  
- productScale  
- productVendor  
- productDescription  
- quantityInStock  
- buyPrice  
- MSRP  

Business Rules

- Each product has a unique `productCode`.  
- A product may appear in zero, one, or many order line items.  

---

Payments

Each **Payment** represents a payment made by a customer.

Attributes

- customerNumber  
- checkNumber  
- paymentDate  
- amount  

Business Rules

- Each payment is made by exactly one customer.  
- A customer may make multiple payments.  
- The combination of `customerNumber` and `checkNumber` uniquely identifies a payment.
